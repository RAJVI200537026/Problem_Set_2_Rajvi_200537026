```python
#question 3
# Define the Marsupial class
class Marsupial:
    def __init__(self):
        # Initialize an empty pouch list
        self.pouch = []

    def put_in_pouch(self, item):
        # Add the item to the pouch list
        self.pouch.append(item)

    def pouch_contents(self):
        # Return the contents of the pouch list
        return self.pouch


# Define the Kangaroo class as a subclass of Marsupial
class Kangaroo(Marsupial):
    def __init__(self, x, y):
        # Initialize the Kangaroo object with coordinates x and y
        Marsupial.__init__(self)  # Call the Marsupial's constructor
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        # Move the Kangaroo by dx units along the x-axis and dy units along the y-axis
        self.x += dx
        self.y += dy

    def __str__(self):
        # Return a string representation of the Kangaroo's coordinates
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"


# Create an instance of Marsupial and test its methods
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())

# Create an instance of Kangaroo and test its methods
k = Kangaroo(0, 0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)
```

```python
#Question 5
def binary(n):
    if n > 1:
        binary(n // 2)  # Recursively call binary() with the quotient of n divided by 2
    print(n % 2, end='')  # Print the binary digit (remainder of n divided by 2)

# Test cases
binary(0)  # Prints 0
print()
binary(1)  # Prints 1
print()
binary(3)  # Prints 11
print()
binary(9)  # Prints 1001
print()
```

```python
#Question 6
from bs4 import BeautifulSoup

class HeadingParser:
    def __init__(self):
        self.headings = []

    def parse_html(self, html_content):
        # Create a BeautifulSoup object by parsing the HTML content
        soup = BeautifulSoup(html_content, 'html.parser')
        
        # Find all heading tags (h1 to h6) in the HTML document
        heading_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
        
        # Iterate over each heading tag
        for tag in heading_tags:
            # Determine the indentation level based on the tag name
            indentation = int(tag.name[1:]) - 1
            
            # Get the text content of the heading tag
            heading_text = tag.get_text().strip()
            
            # Add the heading to the list with the appropriate indentation
            self.headings.append(' ' * indentation + heading_text)

    def print_headings(self):
        # Iterate over each heading in the list and print it
        for heading in self.headings:
            print(heading)

with open('w3c.html', 'r') as infile:
    content = infile.read()

hp = HeadingParser()
hp.parse_html(content)
hp.print_headings() # Print the extracted headings
```

```python
#Question 7
import requests

def webdir(url, depth, indent):
    # Print the URL with the specified indentation
    print(' ' * indent + url)

    # Base case: reached the maximum depth
    if depth == 0:
        return

    try:
        # Send a GET request to retrieve the web page
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            # Extract the URLs from the web page
            urls = extract_urls(response.text)

            # Recursively visit each URL within the specified depth and increased indentation
            for next_url in urls:
                webdir(next_url, depth - 1, indent + 1)
    except requests.RequestException as e:
        print('Error:', e)

def extract_urls(html):
    # This is a simple implementation that uses a regular expression
    # You can use a more robust HTML parsing library like BeautifulSoup for real-world scenarios
    import re
    pattern = r'<a\s+(?:[^>]*?\s+)?href="([^"]*)"'
    return re.findall(pattern, html)

# Example usage
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)


```

```python
#Question 8
import sqlite3

# Connect to the database
conn = sqlite3.connect('rajvi_database.db')
cursor = conn.cursor()

# Create the World_weather_data table if it doesn't exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS weather_data (
        City TEXT,
        Country TEXT,
        Season TEXT,
        Temperature REAL,
        Rainfall REAL
    )
''')

# Insert data into the World_weather_data table
cursor.executemany('''
    INSERT INTO weather_data VALUES (?, ?, ?, ?, ?)
''', [
    ('Mumbai', 'India', 'Winter', 24.8, 5.9),
    ('Mumbai', 'India', 'Spring', 28.4, 16.2),
    ('Mumbai', 'India', 'Summer', 27.9, 1549.4),
    ('Mumbai', 'India', 'Fall', 27.6, 346.0),
    ('London', 'United Kingdom', 'Winter', 4.2, 207.7),
    ('London', 'United Kingdom', 'Spring', 8.3, 169.6),
    ('London', 'United Kingdom', 'Summer', 15.7, 157.0),
    ('London', 'United Kingdom', 'Fall', 10.4, 218.5),
    ('Cairo', 'Egypt', 'Winter', 13.6, 16.5),
    ('Cairo', 'Egypt', 'Spring', 20.7, 6.5),
    ('Cairo', 'Egypt', 'Summer', 27.7, 0.1),
    ('Cairo', 'Egypt', 'Fall', 22.2, 4.5)
])

# Commit the changes to the database
conn.commit()

# Query: All the temperature data
print("Temperature Data:")
cursor.execute("SELECT Temperature FROM weather_data")
temperature_data = cursor.fetchall()
for temperature in temperature_data:
    print(temperature[0])

# Query: All the cities, but without repetition
print("Cities:")
cursor.execute("SELECT DISTINCT City FROM weather_data")
cities = cursor.fetchall()
for city in cities:
    print(city[0])

# Query: All the records for India
print("Records for India:")
cursor.execute("SELECT * FROM weather_data WHERE Country = 'India'")
india_records = cursor.fetchall()
for record in india_records:
    print(record)

# Query: All the Fall records
print("Fall Records:")
cursor.execute("SELECT * FROM weather_data WHERE Season = 'Fall'")
fall_records = cursor.fetchall()
for record in fall_records:
    print(record)

# Query: The city, country, and season for which the average rainfall is between 200 and 400 millimeters
print("Average Rainfall between 200 and 400:")
cursor.execute("SELECT City, Country, Season FROM weather_data WHERE Rainfall BETWEEN 200 AND 400")
rainfall_records = cursor.fetchall()
for record in rainfall_records:
    print(record)

# Query: The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order
print("Average Fall Temperature above 20:")
cursor.execute("SELECT City, Country FROM weather_data WHERE Season = 'Fall' AND Temperature > 20 ORDER BY Temperature ASC")
fall_temperature_records = cursor.fetchall()
for record in fall_temperature_records:
    print(record)

# Query: The total annual rainfall for Cairo
print("Total Annual Rainfall for Cairo:")
cursor.execute("SELECT SUM(Rainfall) FROM weather_data WHERE City = 'Cairo'")
total_rainfall = cursor.fetchone()
print(total_rainfall[0])

# Query: The total rainfall for each season
print("Total Rainfall for Each Season:")
cursor.execute("SELECT Season, SUM(Rainfall) FROM weather_data GROUP BY Season")
total_rainfall_per_season = cursor.fetchall()
for record in total_rainfall_per_season:
    print(record[0], record[1])

# Close the database connection
conn.close()
```

```python
#Question 9
# Define the list of words
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

#a) ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE','LAZY', 'DOG'] uppercase
uppercase_words = [word.upper() for word in words]

#b) ['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the','lazy', 'dog'] lowercase
lowercase_words = [word.lower() for word in words]

# c) [3, 5, 5, 3, 5, 4, 3, 4, 3] (the list of lengths of words in list words).
word_lengths = [len(word) for word in words]

# d) [['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN','brown', 5], ['FOX', 'fox', 3], ['JUMPS', 'jumps', 5], ['OVER', 'over', 4], ['THE', 'the', 3], ['LAZY', 'lazy',4], ['DOG', 'dog', 3]] (the list containing a list for every word of list
#words, where each list contains the word in uppercase and lowercase and the
#length of the word.)
word_info = [[word.upper(), word.lower(), len(word)] for word in words]

# e) ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog'] (the list of words in list words containing 4 or morecharacters.)
long_words = [word for word in words if len(word) >= 4]

# Print the results
print(f"a:-) {uppercase_words}")
print(f"b:-) {lowercase_words}")
print(f"c:-) {word_lengths}")
print(f"d:-) {word_info}")
print(f"e:-) {long_words}")
```
